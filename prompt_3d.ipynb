```python
import torch
from transformers import AutoTokenizer, LlamaForCausalLM
import bitsandbytes, flash_attn

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('NousResearch/Hermes-3-Llama-3.1-8B', trust_remote_code=True)

# Load the model with specific configurations
model = LlamaForCausalLM.from_pretrained(
    "NousResearch/Hermes-3-Llama-3.1-8B",
    torch_dtype=torch.float16,  # Use mixed precision to save memory
    device_map="auto",  # Automatically map model to available devices (e.g., GPU)
    load_in_8bit=False,  # Don't use 8-bit loading
    load_in_4bit=True,  # Use 4-bit quantization for lower memory usage
    use_flash_attention_2=True  # Enable flash attention for faster inference
)
prompts = [
    """system
You are a knowledgeable assistant helping customers make purchasing decisions.
user
Suggest a good mobile phone to buy under ₹20,000 in India.
assistant""",
]

for chat in prompts:
    print(chat)
    input_ids = tokenizer(chat, return_tensors="pt").input_ids.to("cuda")
    generated_ids = model.generate(input_ids, max_new_tokens=150, temperature=0.8, repetition_penalty=1.1, do_sample=True, eos_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(generated_ids[0][input_ids.shape[-1]:], skip_special_tokens=True, clean_up_tokenization_space=True)
    print(f"Response: {response}")
```

```python
import gradio as gr

# Define a function that takes a user prompt and returns product links and a try-on button
def get_products(prompt):
    # For simplicity, we'll just return some dummy data based on the prompt
    products = {
        "shoes": ["http://example.com/shoe1", "http://example.com/shoe2"],
        "shirts": ["http://example.com/shirt1", "http://example.com/shirt2"],
        "hats": ["http://example.com/hat1", "http://example.com/hat2"]
    }

    # Generate links based on the prompt
    links = products.get(prompt.lower(), [])

    # Create try-on button functionality (dummy implementation)
    try_on_button = "Try On" if links else None

    return links, try_on_button

# Define the Gradio interface
def gradio_interface(prompt):
    links, try_on_button = get_products(prompt)

    # If there are links, create a list of clickable links
    if links:
        links_html = "<br>".join(f'<a href="{link}" target="_blank">{link}</a>' for link in links)
        result = f"{links_html}<br><br><button>{try_on_button}</button>"
    else:
        result = "No products found."

    return result

# Set up the Gradio interface
iface = gr.Interface(
    fn=gradio_interface,
    inputs=gr.Textbox(label="Enter a prompt (e.g., 'shoes', 'shirts', 'hats')"),
    outputs=gr.HTML(),
    title="Product Finder and Try-On",
    description="Enter a product type to get links and a try-on button."
)

# Launch the interface
iface.launch()
```
